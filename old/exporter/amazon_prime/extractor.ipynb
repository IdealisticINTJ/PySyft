```python
import pandas as pd
from bs4 import BeautifulSoup

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Find all the <li> elements with class "v-0HEH"
items = soup.find_all('li', class_='v-0HEH')

# List to store the extracted information for the DataFrame
data = []

# Loop through each item and extract the required information
for item in items:
    title_tag = item.find('a', class_='_1NNx6V Nwyhwn')
    title = title_tag.text.strip() if title_tag else ''
    
    # Find the date from the data-automation-id field
    date_tag = item.find_previous('div', {'data-automation-id': lambda x: x and x.startswith('wh-date-')})
    date = date_tag.text.strip() if date_tag else ''
    
    link = f"https://www.amazon.com{title_tag['href']}" if title_tag and title_tag.has_attr('href') else ''
    
    # Extract all episodes watched within the current TV show block
    episode_items = item.find_all('li', class_='v-0HEH YuulXM')
    for episode_item in episode_items:
        episode_title = episode_item.find('div', class_='S6ogEA').find('p').text.strip()
        data.append({
            'date': date,
            'tv_show_name': title,
            'link_to_primevideo_url': link,
            'episode_title': episode_title
        })

# Convert the list to a pandas DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```
